```python
# Dependencies and Setup
import json as json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
from pprint import pprint
from citipy import citipy # use to determine city based on latitude and longitude

# Output File (CSV)
output_data_file = "cities.csv"

# build api call parameters
from config import api_key  # Import API key
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"  # call weather api go get US customary units
time_limit=1 # used to limit the api call to 1 second per call when exceeding 60 per min
```

## Generate Cities List

```python
# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# List for holding lat_lngs and cities
lat_lngs = []
cities_list = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city_nearest = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city_nearest not in cities_list:
        cities_list.append(city_nearest)

# Print the city count to confirm sufficient count
len(cities_list)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# function to call weather api by city
def call_api(cities_list,time_limit):
    city_ntf_count = 0
    api_count = 0
    api_set = 1
    api_results = []

    print('Beginning Data Retrieval') 
    print('-----------------------------')
    # loop through each city to get weather useing a time limit of 1 call per second
    for city in cities_list:
        query_url = f"{url}appid={api_key}&q={city}&units={units}"
        response = requests.get(query_url)
        error_test_json = response.json()
        time.sleep(time_limit)
        api_count = api_count +1
        
        if api_count > 60:
            api_count = 1
            api_set = api_set +1 
            
        print(f'Processing Record {api_count} of Set {api_set} | {city}')
        # check retrun code from the call to the api
        if error_test_json["cod"] == 200:
            api_results.append(response.json()) 
        else:
            print(f' City not found. Skipping...')
            city_ntf_count = city_ntf_count +1 
            
    print('-----------------------------') 
    print('Data Retrieval Complete')
    print(f'number of cities not found {city_ntf_count}')
    print('-----------------------------')
    
    return api_results
```

```python
# call the weather api function
test_results = call_api(cities_list,time_limit)
```

```python
test_results
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# data mungi the weather data to a dataframe
weather_data = []
for wd_index in test_results:
    wd_city = wd_index['name']
    wd_cloudsiness = wd_index['clouds']['all']
    wd_country = wd_index['sys']['country']
    wd_date = wd_index['dt']
    wd_humidity = wd_index['main']['humidity']
    wd_lat = wd_index['coord']['lat']
    wd_lng = wd_index['coord']['lon']
    wd_temp_max = wd_index['main']['temp_max']
    wd_wind_speed = wd_index['wind']['speed']
    wd_row = [wd_city, wd_cloudsiness, wd_country, wd_date, wd_humidity, wd_lat,wd_lng, wd_temp_max, wd_wind_speed]
    weather_data.append(wd_row)
# build dataframe 
city_weather_df = pd.DataFrame(weather_data)
city_weather_df.columns = ["City","Cloudiness", "Country","Date","Humidity", "Lat","Lng","MaxTemp", "WindSpeed"]
city_weather_df.head()
```

```python
# convert the dataframe to csv and write the file out 
city_weather_df.to_csv(output_data_file, index=False)
```

```python
#plot lat lon to check the randomness of cities
# create  a scatter plot, x axis = lat and y axis = long
plt.scatter(city_weather_df.Lat, city_weather_df.Lng,alpha=0.5,)

# Create a title, x label, and y label for our chart
plt.title(f"City Latitue Longitude ({datetime.now().date()})")
plt.xlabel("Latitude")
plt.ylabel("Longiture")
# Incorporate the other graph properties
plt.grid(True)
plt.show()

# Save Figure
plt.savefig("LatLonBubbleplot.png")
```
